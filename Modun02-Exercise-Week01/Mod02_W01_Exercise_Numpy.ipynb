```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# def gray_lightness(R, G, B)

# Đọc ảnh màu
img = mpimg.imread("/content/dog.jpeg")

# Tách các kênh màu R, G, B
R, G, B = img[:, :, 0], img[:, :, 1], img[:, :, 2]

# # Tạo một figure và các axes
# fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# # Hiển thị ảnh grayscale
# axs[0].imshow(R)
# axs[0].set_title('Ảnh đơn màu R')
# axs[0].axis('off')



# # Hàm chuyển đổi một vector RGB thành giá trị grayscale
def color2grayscale_lightness(vector) -> float:
  max_val = np.apply_along_axis(max, axis = 2, arr = vector)
  min_val = np.apply_along_axis(min, axis = 2, arr = vector)
  return (max_val + min_val)/2

# Hàm chuyển đổi một vector RGB thành giá trị grayscale
def color2grayscale_average(vector) -> float:
    return vector[:,:,0] * 1/3 + vector[:,:,1] * 1/3 + vector[:,:,2] * 1/3

# Hàm chuyển đổi một vector RGB thành giá trị grayscale
def color2grayscale_luminosity(vector) -> float:
    return vector[:,:,0] * 0.21 + vector[:,:,1] * 0.72 + vector[:,:,2] * 0.07


#C12:   # Hàm tính
gray_img_01 = color2grayscale_lightness(img)
print(gray_img_01[0, 0])

#C13:   # Hàm tính
gray_img_02 = color2grayscale_average(img)
print(gray_img_02 [0, 0])

#C14:   # Hàm tính
gray_img_03 = color2grayscale_luminosity (img)
print(gray_img_03 [0, 0])

# # Hiển thị ảnh RGB
# # axs[1].imshow(G,cmap='gray')
# axs[1].imshow(color2grayscale_luminosity(img),cmap = 'gray')
# # axs[1].imshow(G)
# # axs[1].imshow(G, cmap='viridis', interpolation='nearest', alpha=0.8
# axs[1].set_title('Ảnh đơn màu G')
# axs[1].axis('off')

# # Hiển thị ảnh với colormap và nội suy
# axs[2].imshow(B)
# # axs[2].imshow(B, cmap='viridis', interpolation='nearest', alpha=0.8)
# axs[2].set_title('Ảnh đơn màu B')
# axs[2].axis('off')

# # Hiển thị figure
# plt.show()
```

**Câu hỏi 20:** Gọi giá trị trung bình của cột Sales là A.

Tạo ra mảng mới scores chứa các giá trị Good, Average và Bad sao cho:

+ nếu giá trị hiện tại > A => giá trị trong mảng mới là Good,

+                      < A thì sẽ là Bad

+                    và = A sẽ là Average.

Sau đó in ra kết quả scores[7:10]

```python
# pandas
mean_sales = df['Sales'].mean()

def classify_sales(value, mean_value):
    if value > mean_value:
        return 'Good'
    elif value < mean_value:
        return 'Bad'
    else:
        return 'Average'

scores = df['Sales'].apply(lambda x: classify_sales(x, mean_sales))


print(f"Kết quả scores[7:10]: {scores[7:10].values}")
```

```python
# numpy
data_sales = data[:, idx_colume_sales]
mean_sales  = data_sales.mean()

scores = np.where( data_sales > mean_sales, "Good",
                           np.where(data_sales < mean_sales, "Bad", "Average"))

print(f"Kết quả scores[7:10]: {scores[7:10]}")
```

```python
# numpy with lamda
data_sales = data[:, idx_colume_sales]
mean_sales  = data_sales.mean()

# function __main__.<lambda>(x): -> todo
classify = lambda x: "Good" if x > mean_sales else ("Bad" if x < mean_sales else "Average")

# Vectorize this lambda function
vectorized_classify = np.vectorize(classify)

# Apply it to the data_sales array
classification = vectorized_classify(data_sales)
print(f"Kết quả scores[7:10]: {classification[7:10]}")
```

**Câu hỏi 21: ** Gọi giá trị trên cột Sales gần nhất với giá trị trung bình cũng chính cột Sales là A.

Tạo ra mảng mới scores chứa các giá trị Good, Average và Bad sao cho:

+  nếu giá trị hiện tại > A => giá trị trong mảng mới là Good,
+                       < A thì sẽ là Bad và
+                       = A sẽ là Average.

Sau đó in ra kết quả scores[7:10]

```python
# pandas
closest_value_to_mean_sales = df['Sales'].iloc[(df['Sales'] - mean_sales).abs().argmin()]

scores_closest = df['Sales'].apply(lambda x: classify_sales(x, closest_value_to_mean_sales))
print(f"Kết quả scores_closest[7:10]: {scores_closest[7:10].values}")
```

```python
# numpy
data_sales = df['Sales'].to_numpy()
data_sales_mean = data_sales.mean()
idx_closest = np.argmin(np.abs(data_sales - data_sales_mean))

value_closest_to_mean_sales = data_sales[idx_closest]

data_classify_by_closest_value = np.where(data_sales > value_closest_to_mean_sales, 'good',
                                          np.where(data_sales < value_closest_to_mean_sales, 'bad', 'average'))

print(f"Kết quả scores_closest[7:10]: {data_classify_by_closest_value[7:10]}")
```
