```python
# Q2
# 1. Xây dựng class Softmax và softmax_stable 
import torch #_python 3.12.0_-># pip3 install --pre torch torchvision torchaudio --index-url https://download.pytorch.org/whl/nightly/cu118
import torch.nn as nn

class MySoftmax(nn.Module):
    def __init__(self):
      super().__init__()
    
    def forward(self,data):
      x_exp = torch.exp(data)               # data.exp() #
      #Tensor 1 chiều
      total = torch.sum(x_exp, dim = 0, keepdims = True) #x_exp.sum(0, keepdims = True) #x_exp.sum() 
      return x_exp/total

class SoftmaxStable(nn.Module):
  def __init__(self):
    super().__init__()

  def forward(self, x):
    c = torch.max(x, dim = 0)
    x_exp = torch.exp(x - c.values)
    total = torch.sum(x_exp, dim = 0, keepdims=True) #x_exp.sum(0, keepdims=True)
    return x_exp / total  

data = torch.Tensor([1, 2, 3])

#----------TEST---------#
if __name__ == "__main__":
  #Q2:
  data = torch.Tensor([5 , 2, 4])
  my_softmax = MySoftmax()
  output = my_softmax( data )
  assert round(output[-1].item(), 2).as_integer_ratio() == 0.26.as_integer_ratio()
  print(output, "=> Q2 -> Answer B")

  #Q3:
  data = torch . Tensor ([1 , 2, 300000000])
  my_softmax = MySoftmax()
  output = my_softmax(data)
  assert round(output[0].item(), 2).as_integer_ratio() == 0.0.as_integer_ratio()
  print(output, "=> Q3 -> Answer C")
  
  #Q4:
  data = torch.Tensor([1 , 2, 3])
  softmax_stable = SoftmaxStable()
  output = softmax_stable(data)
  assert round(output[-1].item(), 2).as_integer_ratio() == (0.67).as_integer_ratio()
  print(output, "=> Q4 -> Answer B")
  
```

Câu hỏi 5: Một người (person) có thể là student, doctor, hoặc teacher. Một student gồm có name  (string), yob (int) (năm sinh), và grade (string). Các bạn thực hiện viết class Student theo mô tả trên (Các bạn sẽ viết thêm describe() method để print ra tất cả thông tin của object) và kết quả
đầu ra là gì? Chọn đáp án đúng nhất bên dưới.

```python
# Characteristics of OOP
from abc import ABC, abstractmethod

# """ * Thuộc tính: name, yob (year of birth) * Phương thức: __init__"""
class Person(ABC):  
  def __init__(self, name: str, yob: int):
    self._name = name
    self._yob = yob
  
  def getYoB(self):
    return self._yob
  
  @abstractmethod
  def describe(self):
    pass

class Student(Person):
  def __init__(self, name: str, yob: int, grade: str):
    super().__init__(name, yob)
    self._grade = grade

  def describe(self):
    print(f"Student - Name: {self._name} - YoB: {self._yob} - Grade: {self._grade}")

class Teacher(Person):
  def __init__(self, name: str, yob: int, subject: str):
    super().__init__(name, yob)
    self._subject = subject
  
  def describe(self):
    print(f"Teacher - Name: {self._name} - YoB: {self._yob} - Subject: {self._subject}")

class Doctor(Person):
  def __init__(self, name: str, yob: int, specialist: str):
    super().__init__(name, yob)
    self._speciality = specialist

  def describe(self):
    print(f"Doctor - Name: {self._name} - YoB: {self._yob} - Speciality: {self._speciality}")


# Ví dụ về cách sử dụng class thừa kế:
if __name__ == "__main__":
    student1 = Student ( name =" studentZ2023 ", yob = 2011 , grade = "6")
    assert student1._yob == 2011
    student1.describe ()
    print("Q5 -> Answer A")
    

    teacher1 = Teacher(name = "teacherZ2023", yob =1991 , subject = "History")
    assert teacher1._yob == 1991
    teacher1.describe()
    print("Q6 -> Answer B")

    doctor1 = Doctor( name = "doctorZ2023", yob =1981 , specialist ="Endocrinologists")
    assert doctor1._yob == 1981
    doctor1.describe()
    print("Q7 -> Answer A")
```

người trong Ward. Viết 
1) add_person(person) method trong Ward class để add thêm một người mới với nghề nghiệp bất kỳ (student, teacher, doctor) vào danh sách người của ward. Tạo ra một ward object, và thêm vào 1 student, 2 teacher, và 2 doctor. 
2) Thực hiện describe() method để in ra tên ward (name) và toàn bộ thông tin của từng người trong ward.

```python
#Q8:
class Ward:
    def __init__(self, name:str):
        self.__name = name
        self.__list_people = list()
    
    def add_person(self, person: Person):
        self.__list_people.append(person)

    def descibe(self):
        print(f"Ward Name: {self.__name}")
        for p in self.__list_people:
            p.descibe()
        
    def count_doctor(self):
        count = sum(1 for person in self.__list_people if isinstance(person, Doctor))
        return count
    
student1 = Student(name ="studentA", yob =2010 , grade ="7")
teacher1 = Teacher(name ="teacherA", yob =1969 , subject   ="Math")
teacher2 = Teacher(name ="teacherB", yob =1995 , subject   ="History")
doctor1  = Doctor(name  ="doctorA", yob =1945 , specialist ="Endocrinologists")

doctor2 = Doctor ( name ="doctorB", yob =1975 , specialist ="Cardiologists")
ward1 = Ward(name ="Ward1")

ward1.add_person(student1)
ward1.add_person(teacher1)
ward1.add_person(teacher2)
ward1.add_person(doctor1)
assert ward1.count_doctor() == 1

ward1.add_person(doctor2)
ward1.count_doctor()

print("Q8 -> Answer D")
```

```python
#Q11+Q12:
class MyQueue: # List -> enqueue(list_end) - dequeu(list_start)
    def __init__(self, capacity = 10):
        # Khởi tạo một queue với kích thước giới hạn.
        self._capacity = capacity
        self._queue = list()

    def is_full(self) -> bool:
        return self._capacity == len(self._queue)

    def is_empty(self) -> bool:
        return len(self._queue) == 0

    def enqueue(self, item):
        # Thêm một phần tử vào hàng đợi.
        if self.is_full():
          raise ValueError("Overflow") #Exception("Overflow")
        self._queue.append(item) 

    def dequeue(self):
        if self.is_empty():
          raise ValueError("Underflow")
        return self._queue.pop(0)

    def size(self):
        return len(self._queue)
    
    def front(self):
        if self.is_empty():
          raise ValueError("Underflow")
        return self._queue[0]


# Ví dụ về cách sử dụng class MyQueue:
if __name__ == "__main__":
    my_queue = MyQueue(capacity = 5)
    my_queue.enqueue(1)
    my_queue.enqueue(2)
    my_queue.enqueue(3)
    print("Size of queue after enqueues:", my_queue.size())
    print("Dequeue:", my_queue.dequeue())
    print("Is queue empty?", my_queue.is_empty())
    print("Dequeue:", my_queue.dequeue())
    print("Is queue empty?", my_queue.is_empty(), end ="\n")

    #Q11:
    print("\nCâu hỏi 11: Answer A")
    queue1 = MyQueue(capacity = 5)
    queue1.enqueue (1)
    assert queue1.is_full() == False
    queue1.enqueue(2)
    print(queue1.is_full())

    #Q11:
    print("\nCâu hỏi 12: Answer D")
    queue1 = MyQueue(capacity = 5)
    queue1.enqueue(1)
    assert queue1.is_full() == False
    queue1.enqueue(2)
    print (queue1.front())
```
